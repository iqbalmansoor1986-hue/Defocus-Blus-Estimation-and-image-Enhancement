```python
############### All filters and Alpha levels with clean image annotations in single frame ###################
from ultralytics import YOLO
import cv2
import numpy as np
import matplotlib.pyplot as plt

# --------------------------
# Load YOLOv8 Nano Detector
# --------------------------
model = YOLO("yolov8n.pt")
model.conf = 0.5
model.iou = 0.45

# --------------------------
# Load Image
# --------------------------
def load_image(image_path):
    return cv2.imread(image_path)

# --------------------------
# Gaussian Blur
# --------------------------
def apply_blur(image, ksize=(11, 11)):
    return cv2.GaussianBlur(image, ksize, 0)

# --------------------------
# Unsharp Mask
# --------------------------
def unsharp_mask(image, alpha=1.5, blur_ksize=(17, 17)):
    blurred_img = cv2.GaussianBlur(image, blur_ksize, 0)
    sharpened = cv2.addWeighted(image, 1 + alpha, blurred_img, -alpha, 0)
    return sharpened

# --------------------------
# Laplacian Boost
# --------------------------
def laplacian_boost(image, weight=0.8):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    lap = cv2.Laplacian(gray, cv2.CV_64F)
    lap = cv2.convertScaleAbs(lap)
    lap_colored = cv2.cvtColor(lap, cv2.COLOR_GRAY2BGR)
    boosted = cv2.addWeighted(image, 1.0, lap_colored, weight, 0)
    return boosted

# --------------------------
# Proposed Enhancement
# --------------------------
def enhance_proposed_pipeline(image, alpha=1.6, laplacian_weight=0.8, blur_kernel=(17,17)):
    sharpened = unsharp_mask(image, alpha=alpha, blur_ksize=blur_kernel)
    enhanced = laplacian_boost(sharpened, weight=laplacian_weight)
    return enhanced

# --------------------------
# Run YOLO Detection
# --------------------------
def detect_and_annotate(image):
    results = model(image)[0]
    return results.plot()

# --------------------------
# Main Pipeline: Grid into One Figure
# --------------------------
def blur_alpha_comparison(image_inputs, output_path="comparison_grid_fixed.png"):
    if isinstance(image_inputs, str):
        image_inputs = [image_inputs]

    kernels = [(5, 5), (11, 11), (17, 17)]
    alphas = [0.5, 1.0, 1.5]

    num_images = len(image_inputs)
    rows_per_image = len(kernels)
    total_rows = num_images * rows_per_image
    cols = len(alphas) + 2  # clean + blurred + enhancements

    fig, axes = plt.subplots(
        total_rows, cols,
        figsize=(20, 2.5 * total_rows),
        gridspec_kw={'hspace': 0.2, 'wspace': 0.05}
    )

    col_titles = ["Clean (Original)", "Blurred", "Enhanced α=0.5", "Enhanced α=1.0", "Enhanced α=1.5"]

    for img_idx, image_path in enumerate(image_inputs):
        original = load_image(image_path)

        for row_idx, kernel in enumerate(kernels):
            global_row_idx = img_idx * rows_per_image + row_idx

            annotated_clean = detect_and_annotate(original)
            blurred_img = apply_blur(original, ksize=kernel)
            annotated_blur = detect_and_annotate(blurred_img)

            # Clean image column — keep axis for label
            axes[global_row_idx, 0].imshow(cv2.cvtColor(annotated_clean, cv2.COLOR_BGR2RGB))
            axes[global_row_idx, 0].tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
            if global_row_idx == 0:
                axes[global_row_idx, 0].set_title(col_titles[0], fontsize=14, weight='bold')

            # Set Gaussian filter label on first column
            axes[global_row_idx, 0].set_ylabel(
                f"Kernel {kernel[0]}×{kernel[1]}",
                fontsize=12, weight='bold', rotation=90, labelpad=12
            )

            # Blurred image column
            axes[global_row_idx, 1].imshow(cv2.cvtColor(annotated_blur, cv2.COLOR_BGR2RGB))
            axes[global_row_idx, 1].axis("off")
            if global_row_idx == 0:
                axes[global_row_idx, 1].set_title(col_titles[1], fontsize=14, weight='bold')

            # Enhanced columns
            for col_idx, alpha in enumerate(alphas):
                enhanced_img = enhance_proposed_pipeline(blurred_img, alpha=alpha, blur_kernel=kernel)
                annotated_enh = detect_and_annotate(enhanced_img)
                axes[global_row_idx, col_idx + 2].imshow(cv2.cvtColor(annotated_enh, cv2.COLOR_BGR2RGB))
                axes[global_row_idx, col_idx + 2].axis("off")
                if global_row_idx == 0:
                    axes[global_row_idx, col_idx + 2].set_title(col_titles[col_idx + 2], fontsize=14, weight='bold')

    plt.tight_layout()
    plt.savefig(output_path, dpi=300)
    plt.show()
    print(f"[INFO] Comparison figure saved at {output_path}")


# --------------------------
# Run
# --------------------------
image_paths = [
    #"/content/1.png",
    #"/content/2.png",
    "/content/3.png",
    "/content/4.png",
    #"/content/5.png"
]

blur_alpha_comparison(image_paths)
```

```python
from ultralytics import YOLO
import cv2
import numpy as np
import matplotlib.pyplot as plt

# --------------------------
# Load YOLOv8 Nano Detector
# --------------------------
model = YOLO("yolov8n.pt")
model.conf = 0.5   # Confidence threshold
model.iou = 0.45   # NMS IoU threshold

# --------------------------
# Laplacian-Based Edge Boosting
# --------------------------
def laplacian_boost(image, weight=0.6):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    lap = cv2.Laplacian(gray, cv2.CV_64F)
    lap = cv2.convertScaleAbs(lap)
    lap_colored = cv2.cvtColor(lap, cv2.COLOR_GRAY2BGR)
    boosted = cv2.addWeighted(image, 1.0, lap_colored, weight, 0)
    return boosted

# --------------------------
# Unsharp Masking Function
# --------------------------
def unsharp_mask(image, alpha=1.5, blur_ksize=(13, 13)):
    blurred_img = cv2.GaussianBlur(image, blur_ksize, 0)
    sharpened = cv2.addWeighted(image, 1 + alpha, blurred_img, -alpha, 0)
    return sharpened

# --------------------------
# Enhancement Pipeline: Unsharp + Laplacian
# --------------------------
def enhance_proposed_pipeline(image, alpha=1.5, lap_weight=0.6):
    sharpened = unsharp_mask(image, alpha=alpha)
    enhanced = laplacian_boost(sharpened, weight=lap_weight)
    return enhanced

# --------------------------
# Inference and Annotation
# --------------------------
def detect_and_annotate(image):
    results = model(image)[0]
    return results.plot()

# --------------------------
# Run on Pre-Blurred Image
# --------------------------
def process_blurred_image(image_path):
    blurred_input = cv2.imread(image_path)
    enhanced = enhance_proposed_pipeline(blurred_input)

    pred_blurred = detect_and_annotate(blurred_input)
    pred_enhanced = detect_and_annotate(enhanced)

    # Display results
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))
    titles = ['Blurred Input', 'Enhanced by Proposed System']
    images = [pred_blurred, pred_enhanced]

    for i in range(2):
        axs[i].imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
        axs[i].set_title(titles[i], fontsize=13)
        axs[i].axis('off')

    plt.tight_layout()
    plt.savefig("blurred_vs_enhanced_comparison.png", dpi=300)
    plt.show()

# --------------------------
# Example Use
# --------------------------
image_path = '/content/1.png'  # Replace with your actual blurred image path
process_blurred_image(image_path)
```

**Calculation of Blur Level and Suggested Alpha Level**

```python
import cv2
import numpy as np
from ultralytics import YOLO

# ---------------------------
# STEP 1: Blur estimation
# ---------------------------
def estimate_blur_level(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    fm = cv2.Laplacian(gray, cv2.CV_64F).var()  # Variance of Laplacian
    return fm

def map_blur_to_kernel_and_alpha(fm):
    # Threshold mapping (tune if needed)
    if fm > 200:
        return (5, 0.5)   # Mild blur
    elif fm > 100:
        return (11, 1.0)  # Moderate blur
    else:
        return (17, 1.5)  # Severe blur

# ---------------------------
# STEP 2: Sharpening function
# ---------------------------
def unsharp_mask(image, alpha):
    blurred = cv2.GaussianBlur(image, (0, 0), 3)
    sharpened = cv2.addWeighted(image, 1 + alpha, blurred, -alpha, 0)
    return sharpened

# ---------------------------
# STEP 3: Annotate image using YOLO
# ---------------------------
def annotate_image(model, image, save_path):
    results = model(image)
    annotated_frame = results[0].plot()
    cv2.imwrite(save_path, annotated_frame)

# ---------------------------
# MAIN PIPELINE
# ---------------------------
if __name__ == "__main__":
    # Load image
    image_path = "/content/1.png"  # <-- Replace with your blurred image path
    image = cv2.imread(image_path)

    # Load YOLOv8n model
    model = YOLO("yolov8n.pt")

    # Step 1: Estimate blur level
    fm = estimate_blur_level(image)
    kernel_size, alpha = map_blur_to_kernel_and_alpha(fm)
    print(f"[INFO] Estimated blur measure: {fm:.2f}")
    print(f"[INFO] Mapped to Gaussian kernel: {kernel_size}x{kernel_size}, Alpha: {alpha}")

    # Step 2: Annotate original blurred image
    annotate_image(model, image, "annotated_blurred.jpg")

    # Step 3: Apply sharpening with chosen alpha
    enhanced_image = unsharp_mask(image, alpha)

    # Step 4: Annotate enhanced image
    annotate_image(model, enhanced_image, "annotated_enhanced.jpg")

    print("[INFO] Processing complete. Saved:")
    print(" - annotated_blurred.jpg")
    print(" - annotated_enhanced.jpg")
```
